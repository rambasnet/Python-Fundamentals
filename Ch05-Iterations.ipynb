### continue loop
- **continue** statement skips the rest of the current loop when continue is executed

````python
# for loop example - continue
for i in range(1, 11): # range ... 1...10
    if i%2 == 0:
        continue
    print(i)
````

### break loop
- **break** statement is used to break a loop
- breaks or exits the loop immidiately
- rest of the current loop and other loops will not be executed!

````python
# for loop example - break
for i in range(10):
    if i == 10:
        break
    print(i)

print('done')
````

## 5.4 for ... else
- you can write an optional else statement with for loop
- the else clause executes when the loop completes normally
    - this means that the loop did not encounter any **break** statement

````python
for i in range(5):
    if i == 2:
        continue
    print(i)
else:
    print('end!')
````

````python
for i in range(5):
    if i == 2:
        break
    print(i)
else:
    print('end!')
````

````python
# for... else example
# write a program to test whether a given number is prime
n = 97
#isPrime = True
for i in range(2, n//2+1):
    if n%i == 0:
        #isPrime = False
        print(n, ' is not prime')
        break
else:
    print(n, 'is prime')
````

## 5.5 while loop 
- while loop is used when you're not sure how many times the loop should be executed
- may execute 0 or many times!
- for loop and while loop can be used interchangeably
- syntax:
``` python
while condition:
    # loop body
```

````python
# infinite loop - will never stop
while True:
    print('hello')
````

````python
i = 1
while i <= 5:
    print(i, 'Hello World')
    i += 1
````

### countdown
- run countdown as a script
- time.sleep(secs)
- os.system('clear'); os.name ==  'posix', 'nt', 'mac', 'os2', 'ce', 'java', 'riscos'

````python
# while loop - countdown
import time
import os

def clearScreen():
    if os.name == 'nt':
        os.system('cls')
    else:
        os.system('clear')

i = 10
while i >= 1:
    print(i)
    time.sleep(1) # sleep for 1 second
    clearScreen()
    i -= 1
    
#time.sleep(1)
print('Hapy new year!')
````

Exercise 2. Write a program to count the number of digits 0 or 5 in a positive ingeger.

````python
# solution 1
# converst positive integer into string and parse one character at a time
n = 12345505
n = str(n)
count0 = 0
count5 = 0
for c in n:
    if c == '0':
        count0 += 1
    elif c == '5':
        count5 += 1
print('There are {} zero(s) and {} five(s) digits in {}.'.format(count0, count5, n))
````

````python
# solution 2
# repeated divide the positive integer by 10 counting the value of remainder until the number is greater than 0
n = 12345505
savedN = n
count0 = 0
count5 = 0
while n > 0:
    n, rem = divmod(n, 10)
    if rem == 0:
        count0 += 1
    elif rem == 5:
        count5 += 1
    
print('There are {} zero(s) and {} five(s) digits in {}.'.format(count0, count5, savedN))
````
