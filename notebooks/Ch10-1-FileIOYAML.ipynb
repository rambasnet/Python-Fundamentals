## write text data to a file
- 3-step process

1. open file with a name in write 'w' or 'a' mode
2. write data
3. close file

````python
# old school - not preferred!!
fw = open('test1.txt', 'w') # w is for write mode
fw.write('words\n=====\n')
fw.write('apple\nball\ncat\ndog\n')
print(fw.write('zebra\n'))
fw.close() #must close the file to actually write data
# to the secondoary storage
````

````python
help(fw)
````

````python
# newer and better syntax - preferred way!!
alist = [1, 2, 3]
with open('words.txt', 'w') as fout:
    fout.write('apple\nball\ncat\ndog\n')
    fout.write('elephant\n')
    fout.write('zebra\n')
    fout.write(str(1))
    fout.write('\n')
    fout.write(str(alist))
    

# file will be automatically closed when with block is finished executing
# fout.write('test\n') # this will not be written as the file is closed; and throws I/O error
````

## read data line by line
- let's create a file with about 10 integers one per line
- then, read the integer line by line into a list of integers

````python
# create a file with 10 integers
# one integer per line
import random
with open('integers.txt', 'a') as fout:
    for i in range(10):
        num = random.randint(1, 1000)
        fout.write(str(num) + '\n')
````

````python
# read the integer line by line into a list
intList = []
with open('integers.txt', 'r') as fin:
    while True:
        num = fin.readline()
        num = num.strip() # strip \n
        if not num:
            break
        print('num = ', num, type(num))
        intList.append(int(num))
````

````python
print(intList)
````

## YAML Files

- YAML is a human-readable data serialization standard that can be used in conjunction with all programming languages and is often used to write configuration files
- YAML stands for "YAML Ain't Markup Language" or "Yet Another Markup Language"
- YAML files have .yaml or .yml extension
- YAML is a superset of JSON, another popular data serialization format
- YAML is designed to be easily readable by humans
- YAML is case-sensitive
- YAML does not allow the use of tabs while creating YAML files; spaces are used instead
- YAML files can be created using any text editor
- YAML files can have comments; comments in YAML files start with the # symbol
- YAML files can have lists and dictionaries
- YAML files can have multi-line strings

### Write/Update a yaml file
- use yaml.dump() to write a dictionary to a yaml file
- let's encrypt password using hashlib and write to a yaml file

````python
def get_random_salt():
    import random
    import string
    # return a random string letters+digits+punctuations of 16 characters
    letters = string.ascii_letters + string.digits + string.punctuation
    return ''.join(random.choices(letters, k=16))
````

````python
def get_hashed_password(password, salt):
    import hashlib
    round = 100_000
    return hashlib.pbkdf2_hmac('sha256', password.encode('utf-8'), salt.encode('utf-8'), round).hex().upper()
````

````python
salt = get_random_salt()
print(f'{salt = }')
password = 'P@ssw0rd' # prompt user to enter the password
hashed_password = get_hashed_password(password, salt)
print(f'{hashed_password = }')
````

````python
# create a new player
new_user = {}
new_user['name'] = 'Michael Jordan'
new_user['email'] = 'jordan@gmail.com'
new_user['password'] = hashed_password
new_user['salt'] = salt
new_user['played'] = 20
new_user['lost'] = 5
````

````python
# add the new player to the data
data.append(new_user)
print(data)
````

````python
# write the updated data back to the file
with open(file, 'w', encoding='utf-8') as f:
    yaml.dump(data, f)

# check the file to see if the new player is added
````
